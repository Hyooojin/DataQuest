### <p style="color:Gray"># Create a Nutritional Index<p/>

<p style="color:Blue">**5. Create a nutritional index**<p/>

### <p style="color:Gray"># Normalize<p/>
>The columns in the data set use different units. As a result, the range of values varies greatly between columns.It has a different influence on the column value.<br/>
<br/>

>dataset의 열은 서로 다른 단위를 사용하고, 그 결과 값의 column마다 값의 범위는 크게 달라진다. column값에 다른 영향력을 미치게 된다. 

#### <p style="color:Gray">To divide all of the values in a column by that column's  maximum value.<p/> 
This way, all of the columns will range from 0 to 1. 
#### <p style="color:Gray">Series.max()<p/> 

```python
# The largest value in the "Energ_Kcal" column.
max_calories = food_info["Energ_Kcal"].max()
print(max_calories)
```

```python
# Divide the values in "Energ_Kcal" by the largest value.
normalized_calories = food_info["Energ_Kcal"] / max_calories
print(normalized_calories)
```

<p style="color:Blue">**6. Normalize the values in columns.**<p/>

```python
normalized_protein = food_info[
    "Protein_(g)"] / food_info["Protein_(g)"].max()
normalized_protein
```

```python
normalized_fat = food_info[
    "Lipid_Tot_(g)"] / food_info["Lipid_Tot_(g)"].max()
normalized_fat
```

### <p style="color:Gray"># Creating a New Column<p/>
우리는 지금까지 column의 변환 결과인 Series object를 변수로 지정했지만, DataFrame에 New Column으로 추가할 수도 있습니다.
I

```python
iron_grams = food_info["Iron_(mg)"] / 1000
food_info["Iron_(g)"] = iron_grams
print(food_info["Iron_(g)"].head(3))
```

<p style="color:Blue">**7. Add a new column.**<p/>

```python
food_info["Normalized_Protein"] = food_info[
    "Protein_(g)"] / food_info["Protein_(g)"].max()
print(food_info["Normalized_Protein"].head(3))
```

```python
food_info["Normalized_Fat"] = normalized_fat
print(food_info["Normalized_Fat"].head(3))
```

### <p style="color:Gray"># Create a Normalized Nutritional Index<p/>
<p style="color:Blue">**8. Create a nutritional index that uses the normalized fat and protein values, instead of the original values.**<p/>
Score = 2 * (Protein_(g)) - 0.75 * (Lipid_Tot_(g))

```python
food_info["Norm_Nutr_Index"] = food_info[
    "Normalized_Protein"] * 2 + food_info["Normalized_Fat"] * (-0.75)
print(food_info["Norm_Nutr_Index"].head(3))
```

### <p style="color:Gray"># Sorting a DataFrame by a Column<p/>
* A unique identifier column.
* To sort the DataFrame by the some column.
* Which foods rank the highest in the soem column.


#### <p style="color:Gray">sort_values()<p/>
To sort the entire DataFrame.

```python
food_info
```

```python
food_info.sort_values("Sodium_(mg)")
```

```python
# Sorts the DataFrame in-place, rather than returning a new DataFrame .
food_info.sort_values("Sodium_(mg)", inplace=True)
```

```python
# Sorts by descending order, rather than ascending.
food_info.sort_values("Sodium_(mg)", inplace=True, ascending=False)
```

<p style="color:Blue">**9. Sort the food_info DataFrame in-place on the Norm_Nutr_Index column in descending order.**<p/>
To explore which foods rank the highest in the Norm_Nutr_Index column.

```python
food_info.sort_values("Norm_Nutr_Index", inplace=True, ascending=False)
```
